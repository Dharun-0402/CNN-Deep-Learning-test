**4.Find the optimum value of k for k-means clustering using 
the elbow method. Plot the elbow curve**

```python
import sklearn
from sklearn.cluster import KMeans
```

```python
css=[]

for i in range(1,7):
  kmeans=KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10, 
                max_iter=300, 
                tol=0.0001,  
                verbose=0, 
                random_state=None, 
                copy_x=True,  
                algorithm='auto')
  
  kmeans.fit(df)

  css.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,7),css)
```

**5.Find the optimum value of k for k-means clustering using 
the silhouette score method and specify the number of 
observations in each cluster using a bar plot**

```python
import sklearn.metrics as metrics
for i in range(4,8):
    kmeans=KMeans(n_clusters=i,
                  init="k-means++",
                  random_state=200)
    kmeans.fit(df)
    labels=kmeans.labels_
    print ("Silhouette score for k(clusters) = "+str(i)+" is "
           +str(metrics.silhouette_score(df,
                                         labels,
                                         metric="euclidean",
                                         sample_size=1000,
                                         random_state=200)))
import matplotlib.pyplot as plt
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(20,15))
for i in [2, 3, 4, 5]:
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)
    visualizer = SilhouetteVisualizer(km, colors='Paired', ax=ax[q-1][mod])
    visualizer.fit(x)
```

**2.Build a sequential LSTM model to predict positive and 
negative sentiments.**

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
crps= []
for i in range(0, 13871):
    review = re.sub('[^a-zA-Z]', ' ', 
                    df1['text'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    crps.append(review)
```

```python
crps
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cvzr= CountVectorizer(max_features = 1900)
```

```python
X = cvzr.fit_transform(crps).toarray()
y = df1['sentiment'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
clsfr= GaussianNB()
clsfr.fit(X_train, y_train)
```

```python
y_pred = clsfr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
